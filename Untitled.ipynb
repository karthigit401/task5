```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap

# Load the dataset
df = pd.read_csv('US_Accidents_March23.csv')
print("‚úÖ Dataset loaded. Shape:", df.shape)

# Convert time column
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')

# Select relevant columns for EDA
df = df[['ID', 'Start_Time', 'Severity', 'City', 'State', 'Temperature(F)', 'Weather_Condition', 'Humidity(%)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Description', 'Start_Lat', 'Start_Lng']]

# Check for missing data
missing = df.isnull().mean().sort_values(ascending=False)
print("\nüîç Missing Data (Top 5):")
print(missing.head())

# Drop rows with missing lat/lng (needed for map)
df = df.dropna(subset=['Start_Lat', 'Start_Lng'])

# Plot: Number of accidents by state (Top 10)
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='State', order=df['State'].value_counts().iloc[:10].index, palette='viridis')
plt.title("Top 10 States by Number of Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("State")
plt.show()

# Plot: Severity distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Severity', palette='Set2')
plt.title("Accident Severity Distribution")
plt.xlabel("Severity (1 = Low, 4 = High)")
plt.ylabel("Count")
plt.show()

# Plot: Weather conditions (Top 10)
plt.figure(figsize=(12, 6))
top_weather = df['Weather_Condition'].value_counts().nlargest(10)
sns.barplot(x=top_weather.index, y=top_weather.values, palette='magma')
plt.title("Top 10 Weather Conditions During Accidents")
plt.xticks(rotation=45)
plt.ylabel("Number of Accidents")
plt.show()

# Create a heatmap of accident locations (limited to avoid performance issues)
sample_df = df[['Start_Lat', 'Start_Lng']].dropna().sample(n=5000, random_state=42)

m = folium.Map(location=[37.0902, -95.7129], zoom_start=5)  # Center of USA
HeatMap(data=sample_df[['Start_Lat', 'Start_Lng']].values, radius=6).add_to(m)
m.save("us_accident_heatmap.html")

print("‚úÖ Heatmap saved as 'us_accident_heatmap.html'")
```
